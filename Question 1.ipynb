## Unsupervised classification

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
#Avoid disturbing the original data
Z = data.copy()
```

```python
mms = MinMaxScaler()
mms.fit(Z)
transformed = mms.transform(Z)
```

```python
Sum_of_squared_distances = []
K = range(1,10)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(transformed)
    Sum_of_squared_distances.append(km.inertia_)
```

```python
plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k or number of clusters')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
# Elbow Method is uggesting to have 4 clusters

km= KMeans(n_clusters=4, init='k-means++', max_iter=300, n_init=10, random_state=0)
pred_y = km.fit_predict(Z)
```

```python
plt.figure()
plt.scatter(Z.iloc[pred_y == 0, 0], Z.iloc[pred_y == 0, 1], s = 50, c = 'red', label = 'Cluster 1')
plt.scatter(Z.iloc[pred_y == 1, 0], Z.iloc[pred_y == 1, 1], s = 50, c = 'blue', label = 'Cluster 2')
plt.scatter(Z.iloc[pred_y == 2, 0], Z.iloc[pred_y == 2, 1], s = 50, c = 'green', label = 'Cluster 3')
plt.scatter(Z.iloc[pred_y == 3, 0], Z.iloc[pred_y == 3, 1], s = 50, c = 'black', label = 'Cluster 4')


plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], s = 300, c = 'orange', label = 'centroids' )
# _centers_ --> also an attribute that can be accessed.
# s --> size of each point
plt.title("K-means clustering applied to Ingredients")

plt.legend()
plt.show()
```
